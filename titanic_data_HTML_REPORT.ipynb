```python
import tkinter as tk
from tkinter import filedialog, messagebox
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px  # Interactive plots like funnel and pie charts ke liye
from pandas_profiling import ProfileReport
import webbrowser
import nbformat as nbf

class DataAnalyzer:
    def __init__(self, root):
        self.root = root
        self.root.title("Create Analytics Report")  # GUI ka Title
        self.root.geometry("900x750")
        
        
        
        
        header = tk.Label(self.root, text="Create Analytics Report", font=("Arial", 24, "bold"), bg="blue", fg="white")
        header.pack(fill=tk.X)

        # Buttons jo different tasks perform karenge
        self.upload_button = tk.Button(self.root, text="Upload Dataset", command=self.upload_dataset, font=("Arial", 14))
        self.upload_button.pack(pady=10)

        self.remove_button = tk.Button(self.root, text="Remove Selected File", command=self.remove_selected_file, state=tk.DISABLED, font=("Arial", 14))
        self.remove_button.pack(pady=10)

        self.python_report_button = tk.Button(self.root, text="Generate Python Report", command=self.generate_python_report, state=tk.DISABLED, font=("Arial", 14))
        self.python_report_button.pack(pady=10)

        self.html_report_button = tk.Button(self.root, text="Generate HTML Report", command=self.generate_html_report, state=tk.DISABLED, font=("Arial", 14))
        self.html_report_button.pack(pady=10)

        self.summary_button = tk.Button(self.root, text="Summary", command=self.show_summary, state=tk.DISABLED, font=("Arial", 14))
        self.summary_button.pack(pady=10)

        self.view_button = tk.Button(self.root, text="View Dataset", command=self.view_dataset, state=tk.DISABLED, font=("Arial", 14))
        self.view_button.pack(pady=10)

        # Variables to Store Dataset and File Path
        self.dataset = None
        self.file_name = ""  # Dynamically file name use karenge report me

    def upload_dataset(self):
        # File dialog se dataset ko upload karna
        self.file_name = filedialog.askopenfilename(filetypes=[("CSV files", "*.csv"), ("Excel files", "*.xlsx *.xls"), ("JSON files", "*.json")])
        if self.file_name:
            try:
                # File type check karke data load karna
                if self.file_name.endswith(".csv"):
                    self.dataset = pd.read_csv(self.file_name)
                elif self.file_name.endswith(".xlsx") or self.file_name.endswith(".xls"):
                    self.dataset = pd.read_excel(self.file_name)
                elif self.file_name.endswith(".json"):
                    self.dataset = pd.read_json(self.file_name)
                else:
                    raise ValueError("Unsupported file format.")

                # Success message aur buttons enable karna
                messagebox.showinfo("Success", f"Dataset '{os.path.basename(self.file_name)}' uploaded successfully!")
                self.python_report_button.config(state=tk.NORMAL)
                self.html_report_button.config(state=tk.NORMAL)
                self.summary_button.config(state=tk.NORMAL)
                self.view_button.config(state=tk.NORMAL)
                self.remove_button.config(state=tk.NORMAL)
            except Exception as e:
                messagebox.showerror("Error", f"Failed to load dataset.\n{e}")

    def remove_selected_file(self):
        # File ko remove karna aur variables reset karna
        self.file_name = ""
        self.dataset = None
        self.python_report_button.config(state=tk.DISABLED)
        self.html_report_button.config(state=tk.DISABLED)
        self.summary_button.config(state=tk.DISABLED)
        self.view_button.config(state=tk.DISABLED)
        self.remove_button.config(state=tk.DISABLED)
        messagebox.showinfo("File Removed", "Selected file has been removed.")

    def generate_python_report(self):
        # Python (Jupyter) Notebook Report generate karna
        if self.dataset is not None:
            try:
                notebook = nbf.v4.new_notebook()
                code = [
                    "# Automatically Generated Python Analysis Report",
                    "import pandas as pd",
                    "import matplotlib.pyplot as plt",
                    "import seaborn as sns",
                    "import numpy as np",
                    "import plotly.express as px",
                    f"df = pd.read_csv(r'{self.file_name}')",  # File ko dynamically load karna
                    "df.head()",
                    "# Basic Information",
                    "df.info()",
                    "df.describe()",
                    "# Checking for Missing Values",
                    "df.isnull().sum()",
                ]

                # Pie Chart (Categorical column ka proportion dikhane ke liye)
                categorical_cols = self.dataset.select_dtypes(include="object").columns
                if len(categorical_cols) > 0:
                    code.append(f"fig = px.pie(df, names='{categorical_cols[0]}', title='Pie Chart of {categorical_cols[0]}')")
                    code.append("fig.show()")

                # Bar Chart (Ek numeric column ka visualization)
                numeric_cols = self.dataset.select_dtypes(include="number").columns
                if len(numeric_cols) > 0:
                    code.append(f"plt.figure(figsize=(10,6))")
                    code.append(f"df['{numeric_cols[0]}'].value_counts().plot(kind='bar', color='orange')")
                    code.append("plt.title('Bar Chart of {numeric_cols[0]}')")
                    code.append("plt.show()")

                # Scatter Plot (Numeric vs Numeric visualization)
                if len(numeric_cols) > 1:
                    code.append(f"plt.figure(figsize=(10,6))")
                    code.append(f"sns.scatterplot(x=df['{numeric_cols[0]}'], y=df['{numeric_cols[1]}'])")
                    code.append("plt.title('Scatter Plot between {numeric_cols[0]} and {numeric_cols[1]}')")
                    code.append("plt.show()")

                # Area Chart (Time series ya numeric data ke liye)
                if len(numeric_cols) > 0:
                    code.append(f"plt.figure(figsize=(10,6))")
                    code.append(f"df['{numeric_cols[0]}'].plot(kind='area', color='skyblue')")
                    code.append("plt.title('Area Chart of {numeric_cols[0]}')")
                    code.append("plt.show()")

                # Bubble Chart (Two numeric columns aur ek size factor ke saath)
                if len(numeric_cols) > 2:
                    code.append(f"fig = px.scatter(df, x='{numeric_cols[0]}', y='{numeric_cols[1]}', size='{numeric_cols[2]}', color='{numeric_cols[0]}', title='Bubble Chart')")
                    code.append("fig.show()")

                # Radar Chart (Numerical columns ke liye)
                if len(numeric_cols) > 1:
                    code.append(f"categories = {numeric_cols[:5]}")  # First 5 numeric columns
                    code.append(f"values = df[categories].mean().values.tolist()")
                    code.append("""
                        angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
                        values += values[:1]
                        angles += angles[:1]
                        fig = plt.figure(figsize=(6, 6))
                        ax = fig.add_subplot(111, polar=True)
                        ax.fill(angles, values, color='green', alpha=0.25)
                        ax.plot(angles, values, color='green', linewidth=2)
                        ax.set_yticklabels([])
                        ax.set_xticks(angles[:-1])
                        ax.set_xticklabels(categories)
                        ax.set_title('Radar Chart')
                        plt.show()
                    """)

                # Funnel Chart (Categorical data ke liye)
                if len(categorical_cols) > 0:
                    code.append(f"fig = px.funnel(df, x='{categorical_cols[0]}', y='count')")
                    code.append("fig.show()")

                # Dashboard (Multiple charts ko combine karne ke liye)
                code.append("""
                    # Create a Dashboard-style Layout using Subplots
                    fig, axs = plt.subplots(2, 2, figsize=(12, 12))
                    axs[0, 0].pie(df[categorical_cols[0]].value_counts(), labels=df[categorical_cols[0]].value_counts().index, autopct='%1.1f%%')
                    axs[0, 0].set_title('Pie Chart')
                    sns.barplot(x=df[numeric_cols[0]].value_counts().index, y=df[numeric_cols[0]].value_counts().values, ax=axs[0, 1])
                    axs[0, 1].set_title('Bar Chart')
                    sns.scatterplot(x=df[numeric_cols[0]], y=df[numeric_cols[1]], ax=axs[1, 0])
                    axs[1, 0].set_title('Scatter Plot')
                    df[numeric_cols[0]].plot(kind='area', ax=axs[1, 1], color='skyblue')
                    axs[1, 1].set_title('Area Chart')
                    plt.tight_layout()
                    plt.show()
                """)

                # Add code to notebook
                for line in code:
                    notebook.cells.append(nbf.v4.new_code_cell(line))

                # Save the notebook
                output_path = os.path.join(os.getcwd(), f"{os.path.splitext(os.path.basename(self.file_name))[0]}_Python_Report.ipynb")
                with open(output_path, "w") as f:
                    nbf.write(notebook, f)

                # Notify user and open the generated report
                messagebox.showinfo("Success", f"Python Report generated: {output_path}")
                webbrowser.open(f"file://{output_path}")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to generate Python report.\n{e}")

    def generate_html_report(self):
        # HTML Report generate karna using pandas profiling
        if self.dataset is not None:
            try:
                report = ProfileReport(self.dataset, title="Enhanced Data Analysis Report", explorative=True)
                output_path = os.path.join(os.getcwd(), f"{os.path.splitext(os.path.basename(self.file_name))[0]}_HTML_Report.html")
                report.to_file(output_path)
                messagebox.showinfo("Success", f"HTML Report generated: {output_path}")
                webbrowser.open(f"file://{output_path}")
            except Exception as e:
                messagebox.showerror("Error", f"Failed to generate HTML report.\n{e}")

    def show_summary(self):
        # Dataset summary ko display karna (Hinglish me)
        if self.dataset is not None:
            summary = (
                f"Total Rows: {self.dataset.shape[0]}\n"
                f"Total Columns: {self.dataset.shape[1]}\n"
                f"Missing Values: {self.dataset.isnull().sum().sum()}\n"
                "\nColumn-wise Details:\n"
                + str(self.dataset.dtypes)
            )
            messagebox.showinfo("Dataset Summary", summary)

    def view_dataset(self):
        # Dataset ko new window me display karna
        if self.dataset is not None:
            top = tk.Toplevel(self.root)
            top.title("Dataset View")
            text = tk.Text(top, wrap=tk.NONE)
            text.insert(tk.END, self.dataset.head().to_string())
            text.pack(expand=True, fill=tk.BOTH)

# Main Application
if __name__ == "__main__":
    root = tk.Tk()
    app = DataAnalyzer(root)
    root.mainloop()
```
